```python
import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
titanic.groupby(['sex'])[['fare']].mean()
```

```python
titanic.groupby(['sex', 'class'])[['survived']].aggregate('mean')
```

```python
titanic.pivot_table('survived', index='sex', columns='class',observed=False) #same thing
```

```python
age = pd.cut(titanic['age'], [0, 18,100])
titanic.pivot_table('survived',['sex', age], 'class',observed=False)
```

```python
fare = pd.qcut(titanic['fare'], 2)
titanic.pivot_table('survived', ['sex', age], [fare, 'class'])
```

```python
titanic.pivot_table(index='sex', columns=['pclass'],
                    aggfunc={'survived':sum, 'fare':'mean'})
```
